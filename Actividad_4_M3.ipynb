```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importar dataset titanic desde seaborn
titanic = sns.load_dataset("titanic")
titanic.head(2)
```

1.Confeccione un gráfico de barras para visualizar hombres y mujeres que han sobrevivido.

```python
mascara_sobrevivientes = titanic['survived'] == 1
titanic_sobrevivientes = titanic[mascara_sobrevivientes]

ax = sns.countplot(
    x="sex",
    data=titanic,
    hue= 'survived',
    palette="colorblind"
)
ax.set_xticks(np.arange(2),labels=['Mujeres', 'Hombres'])
ax.legend(('No sobreviven', 'Sobrevivientes'), loc='upper right', shadow=True)
# agregar titulo
plt.title("Distribución de pasajeros por sexo")

# agregar etiqueta de valores
plt.xlabel("Sexo")
plt.ylabel("Cantidad de pasajeros")
plt.show()

```

2. Cambiar el color del gráfico.

```python
# cambiar color del grafico anterior
custom_palette = ["#FF0000", "#00FF00", "#0000FF", "#FFFF00"]
ax = sns.countplot(
    x="sex",
    data=titanic,
    hue= 'survived',
    palette=['red','blue']
)
ax.set_xticks(np.arange(2),labels=['Mujeres', 'Hombres'])
ax.legend(('No sobreviven', 'Sobrevivientes'), loc='upper right', shadow=True)
# agregar titulo
plt.title("Distribución de pasajeros por sexo")

# agregar etiqueta de valores
plt.xlabel("Sexo")
plt.ylabel("Cantidad de pasajeros")
plt.show()
```

3. Gráfico de barras agrupados

Visualizar la relación de los pasajeros que sobrevivieron, segregados en clases (primera, segunda y tercera), pero también factorizar en qué ciudad se embarcaron.

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
# grafico lado derecho
sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    hue= 'pclass',
    palette="colorblind", ax = axs[0]
)
# agregar titulo
axs[0].set_title("Pasajeros por sexo y clase")
axs[0].set_xlabel("Clases")
axs[0].set_ylabel("Cantidad de Pasajeros")
axs[0].set_xticks(np.arange(2),labels=['Mujeres', 'Hombres'])
axs[0].legend(('Primera Clase', 'Segunda Clase', 'Tercera Clase'), loc='upper right', shadow=True)

# grafico lado izquierdo
sns.countplot(
    x="sex",
    data=titanic_sobrevivientes,
    hue= 'embark_town',
    palette="colorblind", 
    ax = axs[1],
)
# agregar titulos
axs[1].set_title("Pasajeros por sexo y origen embarque")
axs[1].set_xlabel("Origen de Embarque")
axs[1].set_ylabel("Cantidad de Pasajeros")
axs[1].set_xticks(np.arange(2),labels=['Mujeres', 'Hombres'])
axs[1].legend(('Cherbourg', 'Southampton','Queenstown'), loc='upper right', shadow=True)
plt.show()
```
