2. Separe sus datos en 70/30

```python
# predictores elimina columna target
X = np.array(datos_df.drop(columns='ventas'))
# columna target
y = np.array(datos_df.ventas)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.7, test_size=0.3, random_state=42
)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

3. Obtenga su modelo de Regresión lineal

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model on the training set
model.fit(X_train, y_train)
```

4. Entregue la recta matemática de regresión con las estimaciones
obtenidas

```python
# Evaluate the model on the testing set
score = model.score(X_test, y_test)

print("Model Score:", score)

coefficients = model.coef_
intercept = model.intercept_
display("Coefficients:", coefficients)
display("Intercept:", intercept)
```

```python
# ecuación
print(f"el modelo sigue la ecuación")
print(f"ventas =  {intercept[0]:.4f} + {coefficients[0][0]:.4f}tv + {coefficients[0][1]:.4f}radio + {coefficients[0][2]:.4f}periodico")
```

5. Realice un diagnóstico de los residuos del modelo.


```python
# inspect model residuals
y_pred = model.predict(X_test)
residuals = y_test - y_pred
```

```python
# formatea 1 dimensión
y_test=y_test[:,0]
residuals = residuals[:,0]
```

```python
#sns.scatterplot(x=y_test[:,0], y=residuals[:,0])
sns.scatterplot(x=y_test, y=residuals)
```

```python
# grafico de los residuos rewspecto a los datos de entrenamiento
prediccion_train = model.predict(X_train)
residuos_train   = prediccion_train.flatten() - y_train
residuos_train = residuos_train[:,0]
```

```python
# Inspección visual 
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(9, 8))

axs[0, 0].scatter(y_train, prediccion_train, edgecolors=(0, 0, 0), alpha = 0.4)
axs[0, 0].plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()])
axs[0, 0].set_title('Valor predicho vs valor real')
axs[0, 0].set_xlabel('Real')
axs[0, 0].set_ylabel('Predicción')

axs[0, 1].scatter(list(range(len(y_train))), residuos_train,
                   edgecolors=(0, 0, 0), alpha = 0.4)
axs[0, 1].axhline(y = 0, linestyle = '--', color = 'black', lw=2)
axs[0, 1].set_title('Residuos del modelo')
axs[0, 1].set_xlabel('id')
axs[0, 1].set_ylabel('Residuo')

sns.histplot(
    data    = residuos_train,
    stat    = "density",
    kde     = True,
    line_kws= {'linewidth': 1},
    color   = "firebrick",
    alpha   = 0.3,
    ax      = axs[1, 0]
)

axs[1, 0].set_title('Distribución residuos del modelo')
axs[1, 0].set_xlabel("Residuo")

sm.qqplot(
    residuos_train,
    fit   = True,
    line  = 'q',
    ax    = axs[1, 1], 
    color = 'firebrick',
    alpha = 0.4,
    lw    = 2
)
axs[1, 1].set_title('Q-Q residuos del modelo')
axs[1, 1].tick_params(labelsize = 7)

axs[2, 0].scatter(prediccion_train, residuos_train,
                   edgecolors=(0, 0, 0), alpha = 0.4)
axs[2, 0].axhline(y = 0, linestyle = '--', color = 'black', lw=2)
axs[2, 0].set_title('Residuos del modelo vs predicción')
axs[2, 0].set_xlabel('Predicción')
axs[2, 0].set_ylabel('Residuo')

# Se eliminan los axes vacíos
fig.delaxes(axs[2,1])

fig.tight_layout()
plt.subplots_adjust(top=0.9)
fig.suptitle('Diagnóstico residuos');
```

6. Realice el test de normalidad


```python
# diagnosis of model residuals
from scipy.stats import shapiro
from scipy.stats import normaltest

# Shapiro-Wilk Test
_, p = shapiro(residuals)
print(f'Shapiro-Wilk Test: {p:.3f}')
if p > 0.05:
    print('Residuals look Gaussian (fail to reject H0)')
else:
    print('Residuals do not look Gaussian (reject H0)')

# D’Agostino’s K^2 Test
_, p = normaltest(residuals)
print(f'D’Agostino’s K^2 Test: {p:.3f}')
if p > 0.05:
    print('Residuals look Gaussian (fail to reject H0)')
else:
    print('Residuals do not look Gaussian (reject H0)')
```

7. Obtenga las predicciones del modelo


```python
# target de prueba 
y_true = y_test 

# Mean Squared Error (MSE)
mse = mean_squared_error(y_true, y_pred)
print("Mean Squared Error (MSE):", mse)

# Mean Absolute Deviation (MAD)
mad = mean_absolute_error(y_true, y_pred)
print("Mean Absolute Deviation (MAD):", mad)
```

8. Entregue el error del test

```python
# Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_true, y_pred, squared=False)
print("Root Mean Squared Error (RMSE):", rmse)
```
