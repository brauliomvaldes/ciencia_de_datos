```python
def teclanum(cancion):
    try:
        # valida que no contenga doble # o estaciones en blanco
        if "##" in cancion or " " in cancion:
            raise Exception("Error formato de cancin")
        teclas="ABCDEFG"
        teclado={"C":1,"D":3,"E":5,"F":6,"G":8,"A":10,"B":12}
        ans=[]
        for t in cancion.upper():
            if t in teclas:
                # print("tecla:"+str(teclado[t]))
                ans.append(teclado[t])
            else:
                if t == "#" and len(ans)>0:
                    ## print("nota sostenido #")
                    valor_tecla_anterior = int(ans[len(ans)-1]) + 1
                    ans[len(ans)-1] = valor_tecla_anterior
                else:
                    raise Exception("Error formato de cancin")                    
        return salida(ans)  
    except Exception as e:
        print(e)

        
def salida(array):
        # formato de salida requerido
        texto="ans="
        for indice in range(len(array)):
            texto += " "+str(array[indice])
        return texto
        
cancion="EFGGFEDCCDEED"
teclanum(cancion)
```

```python
# prueba con sostenido #
cancion="E#FGGFEDCCDEED"
teclanum(cancion)
```
