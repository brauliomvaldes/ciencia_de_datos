```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.1f}'.format

datos = {
    'calorias':[420, 380, 390, 490, 300],
    'tiempo':[60, 40, 75, 55, 45],
    }
indices = ['L', 'M', 'X', 'J', 'V']

datos_df = pd.DataFrame(datos, index=indices)
datos_df
```

2. Calcular la media, mediana, y desviación típica de ambas columnas.

```python
print("Calorias")
print(f"Media artimética  : {datos_df.calorias.mean():.1f}")
print(f"Mediana           : {datos_df.calorias.median():.1f}")
print(f"Desviación típica : {datos_df.calorias.std():.1f}")

print("\nTiempos")
print(f"Media aritmética  : {datos_df.tiempo.mean():.1f}")
print(f"Mediana           : {datos_df.tiempo.median():.1f}")
print(f"Desviación        : {datos_df.tiempo.std():.1f}")

print("\nTabulado")
datos_df.apply(['mean','median','std'])
```

3. Añadir otra columna booleana al DataFrame para ver si se ha
cumplido el reto de quemar más de 400 calorías por hora. La nueva
columna debe generarse aplicando una fórmula a las otras columnas.

El DataFrame resultante debe ser el siguiente:

dt2.png

```python
# el tiempo esta dado en minutos por lo que hay que transformar a horas durante el proceso de validación
datos_df['reto'] = (datos_df.calorias / ( datos_df.tiempo / 60) ) > 400
datos_df
```

4. Filtrar el DataFrame y devolver otro DataFrame con las filas pares que cumplan que el número de calorías es mayor de 400.

```python
datos_df[datos_df.calorias > 400].iloc[::2]
```

5. Crear a partir del DataFrame una serie con los porcentajes de días que se ha conseguido el reto y los que no.

```python
print(f"El {(datos_df.reto.value_counts(normalize=True)*100)[0]} % de los días se cumplió el reto")
print(f"El {(datos_df.reto.value_counts(normalize=True)*100)[1]} % de los días no se cumplió el reto")
```

6. Crear un gráfico como el de más abajo que muestre la progresión de las calorías y tiempo durante la semana.

```python
plt.plot(datos_df.index, datos_df.calorias, marker='+')
plt.plot(datos_df.index, datos_df.tiempo, marker='o')
plt.title('Evolución diaria')
plt.legend(['Calorías','Tiempo'])
plt.show()
```
