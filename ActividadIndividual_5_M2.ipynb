- ¿Qué son los datos perdidos en Python?

Se refiere a la existencia de una dato que falta o que tiene un formato no determinado, en el caso de los dataframe, esta situación se presenta muy a menudo, es algo muy normal y este valor
perdido o faltante puede aparecer de distintas en una celda del dataframe como por ejemplo como un signo de interrogación, o N/A, como un 0 o simplemente como una celda en blanco, pero en su mayoría se representa como NaN que se refiere a “no un número”.


- ¿Cómo manipular datos perdidos?

Todo depende del tipo del contexto del problema a resolver y es todo un tema digno de analisis pero principalmente se debe enfocar en hallar una alternativa que apunte a resolver esta indeterminación, por eso se deben analizar cada uno de los casos de manera individual. 

Entre las opciones comunmente aplicadas que se encuentran en la literatura, se encuentran las siguientes:

1.- En primer lugar intentar ir a la fuente de los datos, idialmente verificar con la persona o grupo que recopiló los datos para ver si es posible encontrar el valor faltante.

2.- Otra posibilidad es simplemente eliminar los datos donde se
encuentra ese valor perdido. Acá tienes dos opciones, la primera
sería eliminar la variable faltante o eliminar la fila completa en
donde está se encuentre. Esta opción es conveniente en los casos
que no se tiene muchos datos faltantes y que la relevancia del datosea baja o poco indicente en la muestra. La idea es hacer el menor impacto posible y se debe tener cuidado al momento de eliminar datos.

3.- Otra opción es reemplazar los datos por uno que no produzca un sesgo en los mismos, esto se puede hacer, por ejemplo, calculando el valor promedio de la variable completa, esta es una mejor opción ya que no se desperdician datos, sin embargo, es menos preciso ya que necesitamos reemplazar los datos faltantes con una conjetura que puede ser cierta como puede ser falsa. 

De igual forma, no siempre se puede calcular el valor del promedio si el datos no son numéricos, de ser datos categóricos perdidos, se sustituye el valor faltante por el de la categoría que se repide más.

4.- Y la última opción que tenemos es simplemente dejar los datos
faltantes, así tal cual, como datos perdidos. En ocasiones esta
puede ser una buena decisión.

1. Una Tabla de BINGO

Para una tarde de bingo se usarán tablas generadas con las siguientes reglas:

● Los 5 números en cada columna son aleatorios ordenados no
repetidos de acuerdo al rango asignado por columna.

bingo.png

```python
# importa librerias python a utulizar

import pandas as pd
from random import sample
```

```python
# genera numeros aleatorios ordenados y segun rango
# columna B rango entre 1 - 15
b=sorted(sample(range(1,16),5))
# collumna B rango entre 15- 30
i=sorted(sample(range(16,31),5))
# columna B rango entre 31 - 45
n=sorted(sample(range(31,46),5))
# columna B rango entre 46 - 60
g=sorted(sample(range(46,61),5))
# columna B rango entre 61 - 75
o=sorted(sample(range(61,76),5))    
```

● La casilla central de la tabla tiene valor marcado cero o ‘libre’.

```python
# marca casilla centra con valor cero 0
# corresponde a la columna N
n[2] = 0
```

```python
# prepara diccionario para crear DataFrame
datos={
  'B':b,
  'I':i,
  'N':n,
  'G':g,
  'O':o
}
```

Realice un programa para generar una tabla de bingo.

```python
# crea DataFrame sin mostrar indices
indice=['','','','','']
tabla_de_bingo_df = pd.DataFrame(datos,index=indice)
tabla_de_bingo_df
```

```python
# crea DataFrame mostrando indice numérico
tabla_de_bingo_df = pd.DataFrame(datos)
tabla_de_bingo_df
```
