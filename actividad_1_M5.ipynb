- Aprendizaje supervisado

Busca predecir casos futuros a partir de datos cuya respuesta ya se cocnoce. El algoritmo aprende a partir de las variables explicativas asociadas a la variable respuesta (numérica o categórica): Luego predice el valor  de la variable respuesta cuando se presentan nuevas variables explicativas.
En el Aprendizaje Supervisado los datos para el entrenamiento incluyen la solución deseada, llamada “etiquetas” (labels). 

Un claro ejemplo es al clasificar correo entrante entre Spam o no. Entre las diversas características que queremos entrenar deberemos incluir si es correo basura o no con un 1 o un 0.

- Aprendizaje sin supervisión

Algoritmos que aprenden a partir de la variable explicativa sin ninguna variable respuesta asociada, los datos de entrenamiento no incluyen etiquetas, lo que le permite determinar los patrones de datos y clasificar por sí misma. Los algoritmos se dejan guiar por sus propios mecanismos para descubrir la estructura de datos.

Un ejemplo en el que se usa es para agrupar la información recolectada sobre usuarios en una Web o en una app y que nuestra Inteligencia detecte diversas características que tienen en común.


- Aprendizaje por refuerzo

En aprendizaje por refuerzo (Reinforcement Learning) no tenemos una “etiqueta de salida”, por lo que no es de tipo supervisado y si bien estos algoritmos aprenden por sí mismos, tampoco son de tipo no supervisado, en donde se intenta clasificar grupos teniendo en cuenta alguna distancia entre muestras.

El Reinforcement Learning entonces, intentará hacer aprender a la máquina mediante prueba y error, basándose en un esquema de “premios y castigos”, en un entorno en donde hay que tomar acciones y que está afectado por múltiples variables que cambian con el tiempo.

- Algoritmos de regresión

La programación de este algoritmo es dinámica, pues optimiza los valores de las distribuciones marginales, la metodología consta de dos partes, primero avanza en el tiempo mientras que en la segunda retrocede en el tiempo. Los algoritmos de Regresión modelan la relación entre distintas variables (features) utilizando una medida de error que se intentará minimizar en un proceso iterativo para poder realizar predicciones “lo más acertadas posible”. Se utilizan mucho en el análisis estadístico.

- Algoritmos de agrupación

Se utilizan para agrupar datos existentes de los que desconocemos sus características en común o queremos descubrirlas.
Estos métodos intentan crear “puntos centrales” y jerarquías para diferenciar grupos y descubrir características comunes por cercanía.

La agrupación se centra en la identificación de grupos de registros similares y en el etiquetado de los registros según el grupo al que pertenecen. Esto se hace sin conocimiento previo de los grupos y sus características. Los tipos de algoritmos de agrupación incluyen la agrupación de K-means, K-Medians, Hierarchical Clustering, TwoStep y Kohonen.

- Algoritmos de Aprendizaje Profundo

Basado en redes neuronales artificiales, se llama profundo porque la estructura se compone de varias capas de entrada, salida y ocultas. Cada capa contiene unidades que transforman los datos de entrada en información que la capa siguiente puede usar para realizar una tarea de predicción determinada. Xon las técnicas de feedforward el algoritmo itera de “izquierda a derecha” calculando el valor de activación de las neuronas desde las entradas hasta obtener los valores de salida y con backpropagtion el algoritmo itera para aprender, irá de “derecha a izquierda” en la red para mejorar la precisión de las predicciones, en dos fases, propagar y actualizar pesos.
